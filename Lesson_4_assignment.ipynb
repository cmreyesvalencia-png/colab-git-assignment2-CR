```python
# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers.
# Then, try to add an element to each and observe the difference.

# Your code here:

print("\n=== Task 6: Tuples and Sets ===\n")
try:
    # Create tuple and set
    fruits_tuple = ('apple', 'banana', 'orange')
    numbers_set = {1, 2, 3, 4, 5}

    print(f"Original tuple: {fruits_tuple}")
    print(f"Original set: {numbers_set}")

    # Demonstrate tuple immutability
    print(f"\nTuple ID before operation: {id(fruits_tuple)}")

    try:
        # This creates a NEW tuple, doesn't modify the original
        fruits_tuple += ('grape',)
        print(f"After 'adding' to tuple: {fruits_tuple}")
        print(f"Tuple ID after operation: {id(fruits_tuple)} - NEW OBJECT!")
    except Exception as e:
        print(f"Cannot modify tuple: {e}")

    # Demonstrate set mutability
    print(f"\nSet ID before operation: {id(numbers_set)}")
    numbers_set.add(6)  # Sets are mutable - modifies the original
    print(f"After adding to set: {numbers_set}")
    print(f"Set ID after operation: {id(numbers_set)} - SAME OBJECT!")

    print("\nExplanation: Tuples are immutable - 'adding' creates a new tuple object.")
    print("Sets are mutable - adding modifies the original set object.")

except Exception as e:
    print(f"Error with tuples/sets: {e}")
```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:

print("\n=== Task 9: Iterators and Generators ===")
try:
    # Iterator class
    class EvenNumberIterator:
        def __init__(self, count):
            self.count = count
            self.current = 0

        def __iter__(self):
            return self

        def __next__(self):
            if self.current >= self.count:
                raise StopIteration
            result = self.current * 2
            self.current += 1
            return result

    # Generator function
    def even_number_generator(count):
        for i in range(count):
            yield i * 2

    # Demonstrate iterator
    print("Using iterator:")
    iterator = EvenNumberIterator(5)
    for num in iterator:
        print(num)

    # Demonstrate generator
    print("\nUsing generator:")
    generator = even_number_generator(5)
    for num in generator:
        print(num)
except Exception as e:
    print(f"Error with iterators/generators: {e}")
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

# Your code here:

print("\n=== Task 10: Map, Reduce, and Filter ===\n")
try:
    numbers = [1, 2, 3, 4, 5]

    # Map - square all numbers
    squared = list(map(lambda x: x**2, numbers))
    print(f"Original numbers: {numbers}")
    print(f"Squared numbers (map): {squared}")

    # Reduce - product of all numbers
    product = reduce(lambda x, y: x * y, numbers)
    print(f"Product of all numbers (reduce): {product}")

    # Filter - even numbers only
    evens = list(filter(lambda x: x % 2 == 0, numbers))
    print(f"Even numbers (filter): {evens}")
except Exception as e:
    print(f"Error with map/reduce/filter: {e}")
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:

print("\n=== Task 11: OOP - Rectangle Class ===")
try:
    class Rectangle:
        def __init__(self, length, width):
            if length <= 0 or width <= 0:
                raise ValueError("Length and width must be positive")
            self.length = length
            self.width = width

        def area(self):
            return self.length * self.width

        def perimeter(self):
            return 2 * (self.length + self.width)

        def __str__(self):
            return f"Rectangle({self.length}x{self.width})"

    # Create rectangle objects
    rect1 = Rectangle(5, 3)
    rect2 = Rectangle(7, 4)

    print(f"Rectangle 1: {rect1}")
    print(f"Area: {rect1.area()}, Perimeter: {rect1.perimeter()}")

    print(f"\nRectangle 2: {rect2}")
    print(f"Area: {rect2.area()}, Perimeter: {rect2.perimeter()}")
except Exception as e:
    print(f"Error with OOP: {e}")
```
